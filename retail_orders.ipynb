```python
# Extract the file from zip file
import zipfile
zip_ref = zipfile.ZipFile("orders.csv.zip")
zip_ref.extractall() #extract file to dir
zip_ref.close() # closing the file
```

```python
# Read data from csv file and handle null values
import pandas as pd
df = pd.read_csv("orders.csv",na_values=['Not Available','unknown'])
df.head(10)
```

```python
df['Ship Mode'].unique()
```

```python
# Rename the column names, make them lower case and replace space with '_'
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ",'_')
df.head()
```

```python
df.dtypes
```

```python
# Convert order_date object to datetime format
df['order_date'] = pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
# Derive new columns discount, sale price, profit
df['discount'] = df['list_price'] * df['discount_percent']*0.01
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
df.head()
```

```python
# Drop the columns cost_price, list_price, discount_percent	
df.drop(columns=['cost_price', 'list_price', 'discount_percent'],inplace=True)
```

```python
# connect the database and load it in the sql server
import pyodbc
import sqlalchemy as sa
from sqlalchemy.sql import text

# DSN name created in ODBC Data Source Administrator
dsn = 'sqlexpress'

# Construct the connection string using DSN
connection_string = f"DSN={dsn};Trusted_Connection=yes;"

# Create an engine
engine = sa.create_engine(f'mssql+pyodbc:///?odbc_connect={connection_string}')

conn = engine.connect()
# # Test the connection
# with engine.connect() as conn:
#     result = conn.execute(text("SELECT 1"))
#     print(result.fetchone())
```

```python
# pandas create highest datatype if we create a table using replace
# df.to_sql('df_orders',con=conn,index=False,if_exists='replace')

# always better create a table in sql and append the rows using below code
df.to_sql('df_orders',con=conn,index=False,if_exists='append')
```
